# Q2. Write code that does the following:
* Input : 
```
str_sample = 'a b b b c c d'
```
* Output :
```
>>> { 'a' : 1, 'b' : 3, 'c' : 2, 'd' : 1 }
 ```

````python
####################
## your code here ##
####################
from collections import defaultdict

# Input
str_sample = input("str_sample = ").split()

dic = defaultdict(int)
for i in str_sample:
    dic[i] += 1

# Output
dic = dict(dic)
print(dic)
````

# Q3. Make a list that contains the values of origin(the list below) if only the values is higher than equal to 6

````python
origin = [ 9 ,5,6,2,1,19,34,1,36,9,5,4,162,125,62,1,23,3,6,1,24,4]

####################
## your code here ##
####################

ans = [i for i in origin if i >= 6]
print(ans)
````
